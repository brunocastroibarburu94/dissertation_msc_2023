### Statistics calculations

````julia
using Statistics
````

````julia
# Basic Statistics Figures
m1=mean(df[:,:Close_AAPL]) # Mean of Close price of AAPL
s1=std(df[:,:Close_AAPL]) # Standard deviation of Close price of AAPL
v1=var(df[:,:Close_AAPL]) # Variance deviation of Close price of AAPL
println("Mean Close price for Apple (AAPL) in Jan 2012: $(mean(df[:,:Close_AAPL]))")
println("Standard Deviation of Close price for Apple (AAPL) in Jan 2012: $s1")
println("Variance of Close price for Apple (AAPL) in Jan 2012: $v1")
````
