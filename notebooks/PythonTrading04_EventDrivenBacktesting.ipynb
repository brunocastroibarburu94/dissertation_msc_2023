# Interday backtesting using Zipline-Reloaded.


In the console first we ingest the data from Quandl (note that the environment variable QUANDL_API_KEY, needs to be set up). And then we execute a strategy zipline_demo_strategy.

```bash
zipline ingest -b quandl
zipline run -f zipline_demo_strategy.py --start 2014-1-1 --end 2018-1-1 -o dma.pickle --no-benchmark
zipline run -f zipline_demo_strategy.py --start 2014-1-1 --end 2018-1-1 -o dma2.pickle --no-benchmark
```

The outputs are stored in the file dma.pickle and is read below
